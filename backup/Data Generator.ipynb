## poisson distribution

```python
import pandas as pd 
import numpy as np 
# import matplotlib.pyplot as plt

dates = pd.date_range('20190101', periods=365)
df = pd.DataFrame({'item_id': 'item10',
                   'target_value': np.random.poisson(lam=50, size=365)}, index=dates)
print(df)
# df.to_csv (r'export_dataframe.csv', index = False, header=True)
df.head(50).plot(figsize=(15,10))
```

```python
import pandas as pd 
import numpy as np 
import datetime
import random

start = datetime.datetime.strptime("2019-01-01", "%Y-%m-%d")
end = datetime.datetime.strptime("2019-12-31", "%Y-%m-%d")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days+1)]

data_list = []

# 금요일, 토요일 매출이 많은 경우
for date in date_generated:
    if(date.weekday() == 4 or date.weekday() == 5):
        data_list.append(random.sample(range(200,350),1)[0])
    else:
        data_list.append(random.sample(range(50,100),1)[0])
    

dates = pd.date_range('20190101', periods=365)
df = pd.DataFrame({'item_id': 'item15',
                   'target_value': data_list}, index=dates)
print(df)
# df.to_csv (r'export_dataframe.csv', index = False, header=True)
df.head(50).plot(figsize=(15,10))
    
```

## for iteration specific day

```python
import pandas as pd 
import numpy as np 
import datetime
import random

start = datetime.datetime.strptime("2019-01-01", "%Y-%m-%d")
end = datetime.datetime.strptime("2019-12-31", "%Y-%m-%d")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days+1)]
    

dates = pd.date_range('20190101', periods=365)
df_total=pd.DataFrame()

# 하루 날짜
for i in range(100):
    data_list = []
    
    tg_date = random.randrange(6)

    tg_date_value = random.randrange(200, 500)
    tg_date_value_gap = random.randrange(10, 200)
    
    nm_date_value = random.randrange(150)
    nm_date_value_gap = random.randrange(1,50)
    
    for date in date_generated:
        if(date.weekday() == tg_date):
            data_list.append(random.sample(range(tg_date_value,tg_date_value+tg_date_value_gap),1)[0])
        else:
            data_list.append(random.sample(range(nm_date_value,nm_date_value+nm_date_value_gap),1)[0])
    
    df = pd.DataFrame({'item_id': 'item'+str(1+i),
                       'target_value': data_list}, index=dates)
    
    # df.head(50).plot(figsize=(15,10))
    df_total=df_total.append(df)

# 이틀 날짜 
for i in range(100):
    data_list = []
    
    list = [0,1,2,3,4,5,6]
    tg_date_list = random.sample(list,2)

    tg_date_value = random.randrange(200, 500)
    tg_date_value_gap = random.randrange(10, 200)
    
    nm_date_value = random.randrange(150)
    nm_date_value_gap = random.randrange(1,50)

    for date in date_generated:
        if(date.weekday() == tg_date_list[0] or date.weekday() == tg_date_list[1]):
            data_list.append(random.sample(range(tg_date_value,tg_date_value+tg_date_value_gap),1)[0])
        else:
            data_list.append(random.sample(range(nm_date_value,nm_date_value+nm_date_value_gap),1)[0])

    df = pd.DataFrame({'item_id': 'item'+str(101+i),
                       'target_value': data_list}, index=dates)
    
    # df.head(50).plot(figsize=(15,10))
    df_total=df_total.append(df)

# print(len(df_total))

df_total.to_csv (r'special_day_trend_data.csv', index = True)
# df.head(50).plot(figsize=(15,10))
```

## for iteration steady increasing sales

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import datetime
import random
import boto3

dates = pd.date_range('20190101', periods=365)
df_total=pd.DataFrame()

for i in range(300):
    results = np.arange(365)*random.uniform(0, 0.5)
    noise = np.random.randn(365)*random.randrange(50)
    noise = [abs(i) for i in noise]

    results += noise
    results = [int(i) for i in results]
    
    df = pd.DataFrame({'item_id': 'item'+str(201+i),
                       'target_value': results}, index=dates)
    
    # df.plot(figsize=(15,10))
    df_total=df_total.append(df)
    
# print(df_total)
# print(len(df_total))
# df_total.to_csv (r'steady_increase_trend_data.csv', index = True)


bucketName='forecast-demogo-bucket'
region = boto3.Session().region_name
session = boto3.Session(region_name=region) 
s3 = session.client('s3')

s3.upload_file('data/steady_increase_trend_data.csv', bucketName, 'steady_increase_trend_data.csv')
```
